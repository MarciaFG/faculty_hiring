```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Helvetica']
rcParams['xtick.major.pad']='8'
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import palettable

import networkx as nx
import fnmatch
import scipy as sp

from scipy.io import loadmat
from scipy.stats import ks_2samp
from os import path, walk, system
from sklearn.decomposition import TruncatedSVD
from sklearn.neighbors import KernelDensity
from collections import defaultdict, Counter
from scipy.special import gamma as gammaf

# Local imports
#from rewiring import threshold_matrix, which_edge_eig, which_edge_high, which_edge_product
```

```python
# Where to save figures:
figures_dir = '/Users/samway/Desktop/'

# Absolute path to the HumanAF directory:
data_dir = '/Users/samway/Dropbox/ITHeart/data/mi_matrix/HumanAF/'

# Which method to use for removing edges:
# Select from [which_edge_eig, which_edge_high, which_edge_product]
#which_edge = which_edge_eig
#which_edge_tech = which_edge.__name__.split('_')[-1]  # Which technique?

patients = ['CM031115','CS033015','DC021015','EC012815','EP010815','JC032415','JO041315',
            'LA041615','MD012815','MH021115','MM040615','PD021315','WS020415']

# Patients with prior ablation:
with_pa = ['CM031115', 'EC012815', 'WS020415', 'CS033015', 'JC032415', 'MM040615']

# Exclude the following patients (optional)
leave_out = ['CS033015','MD012815','MH021115','PD021315','WS020415']
leave_out = []  # <-- UNCOMMENT THIS LINE TO LEAVE ALL PATIENTS IN ANALYSIS!

patients_with_pa = [p for p in patients if p not in leave_out and p in with_pa]
patients_without_pa = [p for p in patients if p not in leave_out and p not in with_pa]
all_patients = patients_with_pa + patients_without_pa
num_patients = len(all_patients)
```

```python
# General variables
num_electrodes = 64      # Matrices are 64x64 pairwise mutual information scores between electrodes. 
num_windows = 10         # Number of timepoints, which we're treating as independent samples.

# Color stuff
COLOR_WITH_PA = array([1.0, 0, 0])    #colors[0]
COLOR_WITHOUT_PA = array([0, 0, 1.0]) #colors[1]

# Plotting variables
BAR_WIDTH = 0.6
TICK_SIZE = 14
XLABEL_PAD = 10
LABEL_SIZE = 16
LEGEND_SIZE = 12
LINE_WIDTH = 2

'''
    HumanAF contains a directory for each patient. Each of those directories contains an LA1 and RA1
    subdirectory. These are for the left and right atrium. Insde each subdirectory is a "KSG_MI.mat"
    file, which contains that mutual information matrices ('mi' variable).
    
    The 'mi' matrix is a 64x64x10 data structure. There are ten 64x64 time windows.
'''

# Load left atrium (LA1) data
la_data = {}
for root, dirnames, filenames in walk(data_dir):
    for f in fnmatch.filter(filenames, 'KSG_MI.mat'):
        if root.endswith('LA1'):  # if containing folder is..
            tag = path.basename(root.replace('/LA1', ''))
            file_path = path.join(root, f)
            la_data[tag] = loadmat(file_path)['mi']

# Load right atrium (RA1) data
ra_data = {}
for root, dirnames, filenames in walk(data_dir):
    for f in fnmatch.filter(filenames, 'KSG_MI.mat'):
        if root.endswith('RA1'):  # if containing folder is..
            tag = path.basename(root.replace('/RA1', ''))
            file_path = path.join(root, f)
            ra_data[tag] = loadmat(file_path)['mi']
```

# Grab all MI scores

```python
LA_control = []
RA_control = []
LA_ablation = []
RA_ablation = []

for p in patients_without_pa:
    for w in xrange(num_windows):
        L = la_data[p][:,:,w]
        R = ra_data[p][:,:,w]
        L = np.triu(L,1)  # Grab the upper triangle values -- don't want the diagonal (self-information) and 
        R = np.triu(R,1)
        L[where(L<0)] = 0
        R[where(L<0)] = 0
        L = L[np.nonzero(L)]
        R = R[np.nonzero(R)]
        LA_control += L.tolist()
        RA_control += R.tolist()
        
for p in patients_with_pa:
    for w in xrange(num_windows):
        L = la_data[p][:,:,w]
        R = ra_data[p][:,:,w]
        L = np.triu(L,1)
        R = np.triu(R,1)
        L[where(L<0)] = 0
        R[where(L<0)] = 0
        L = L[np.nonzero(L)]
        R = R[np.nonzero(R)]
        LA_ablation += L.tolist()
        RA_ablation += R.tolist()
```

```python
def hide_right_top_axis(ax):
    """ Remove the top and right axis """
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9,4))

show_kde = True
show_hist = False

# Histogram parameters 
bins = np.linspace(0, 3., 225)
bin_width = diff(bins)[0]

# KDE parameters
x_plot_steps = 1000
x_plot = linspace(-0., 2., x_plot_steps)[:, np.newaxis]
x_plot_step = np.abs(x_plot[1] - x_plot[0])

# LEFT ATRIUM
bw = 2*1.06*std(LA_control)*(len(LA_control)**(-1.0/5))
LA_control = np.array(LA_control)
LA_ablation = np.array(LA_ablation)
if show_hist:
    w = np.ones_like(LA_control)
    w /= np.sum(w)
    axs[0].hist(LA_control, bins, weights=w, normed=False, alpha=0.3, color=COLOR_WITHOUT_PA)
    w = np.ones_like(LA_ablation)
    w /= np.sum(w)
    axs[0].hist(LA_ablation, bins, weights=w, normed=False, alpha=0.3, color=COLOR_WITH_PA)
else:
    bin_width = 1.0 # Don't rescale the KDE

if show_kde:
    data = LA_control[np.where(LA_control > 0)]
    data = data[:, np.newaxis]
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data)
    density = np.exp(kde.score_samples(x_plot)) * bin_width  
    #axs[0].fill_between(x_plot[:, 0], density, np.zeros_like(density), color=COLOR_WITHOUT_PA, alpha=0.2)
    axs[0].plot(x_plot[:, 0], density, c=COLOR_WITHOUT_PA, linewidth=LINE_WIDTH, label='Control group')

    data = LA_ablation[np.where(LA_ablation > 0)]
    data = data[:, np.newaxis]
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data)
    density = np.exp(kde.score_samples(x_plot)) * bin_width  
    #axs[0].fill_between(x_plot[:, 0], density, np.zeros_like(density), color=COLOR_WITH_PA, alpha=0.2)
    axs[0].plot(x_plot[:, 0], density, c=COLOR_WITH_PA, linewidth=LINE_WIDTH, label='Ablation group')

    
# RIGHT ATRIUM
bw = 2*1.06*std(RA_control)*(len(RA_control)**(-1.0/5))
RA_control = np.array(RA_control)
RA_ablation = np.array(RA_ablation)
if show_hist:
    w = np.ones_like(RA_control)
    w /= np.sum(w)
    axs[1].hist(RA_control, bins, weights=w, normed=False, alpha=0.3, color=COLOR_WITHOUT_PA)
    w = np.ones_like(RA_ablation)
    w /= np.sum(w)
    axs[1].hist(RA_ablation, bins, weights=w, normed=False, alpha=0.3, color=COLOR_WITH_PA)

if show_kde:
    data = RA_control[np.where(RA_control > 0)]
    data = data[:, np.newaxis]
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data)
    density = np.exp(kde.score_samples(x_plot)) * bin_width  
    #axs[1].fill_between(x_plot[:, 0], density, np.zeros_like(density), color=COLOR_WITHOUT_PA, alpha=0.2)
    axs[1].plot(x_plot[:, 0], density, c=COLOR_WITHOUT_PA, linewidth=LINE_WIDTH, label='Control group')

    data = RA_ablation[np.where(RA_ablation > 0)]
    data = data[:, np.newaxis]
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data)
    density = np.exp(kde.score_samples(x_plot)) * bin_width  
    #axs[1].fill_between(x_plot[:, 0], density, np.zeros_like(density), color=COLOR_WITH_PA, alpha=0.2)
    axs[1].plot(x_plot[:, 0], density, c=COLOR_WITH_PA, linewidth=LINE_WIDTH, label='Ablation group')
    leg = axs[1].legend(fontsize=LABEL_SIZE-2, frameon=False, loc='center right')

    for legobj in leg.legendHandles:
        legobj.set_linewidth(4.0)

xlims = (0, 2.)
axs[0].set_xlim(xlims)
axs[1].set_xlim(xlims)

if show_hist:
    ylims = (0, 0.04)
    axs[0].set_ylim(ylims)
    axs[1].set_ylim(ylims)
    axs[0].set_ylabel('Probability', fontsize=LABEL_SIZE)
    axs[1].set_ylabel('Probability', fontsize=LABEL_SIZE)
else:
    ymax = np.max([axs[0].get_ylim()[1], axs[1].get_ylim()[1]])
    axs[0].set_ylim([0., ymax])
    axs[1].set_ylim([0., ymax])
    axs[0].set_ylabel('Probability density', fontsize=LABEL_SIZE)
    #axs[1].set_ylabel('Probability density', fontsize=LABEL_SIZE)
    

axs[0].set_xlabel('Mutual Information (nat)', fontsize=LABEL_SIZE)
axs[1].set_xlabel('Mutual Information (nat)', fontsize=LABEL_SIZE)
axs[0].tick_params(axis='both', which='major', labelsize=TICK_SIZE)
axs[1].tick_params(axis='both', which='major', labelsize=TICK_SIZE)
axs[0].set_title('Left Atrium', fontsize=LABEL_SIZE+2, y=1.02)
axs[1].set_title('Right Atrium', fontsize=LABEL_SIZE+2, y=1.02)

hide_right_top_axis(axs[0])
hide_right_top_axis(axs[1])

plt.savefig(figures_dir + 'fig03.pdf')
plt.show()
```
