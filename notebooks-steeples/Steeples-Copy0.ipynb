## Load data

```python
inst = institution_parser.parse_institution_records(open(inst_file))
asst_faculty = load.load_assistant_profs(open(faculty_file), inst)
ignore = ['All others', 'UNKNOWN', 'CUNY Graduate Center']
inst_keys = [place for place in inst if place not in ignore]
inst_keys = [s[1] for s in sorted([(inst[place]['pi'], place) for place in inst_keys])]
inst_keys_top = inst_keys[:99]

# What are the topics?
topic_descriptions = ['robots/hci', 
                      'security', 
                      'theory', 
                      'vision', 
                      'arch', 
                      'distributed', 
                      'data', 
                      'sensor nets', 
                      'software', 
                      'ml']
num_topics = len(topic_descriptions)

# Limit for entropy
max_d = np.ones(num_topics, dtype=float) / num_topics
maximum_entropy = -np.dot(max_d, np.log2(max_d))

def get_asst_location(person):
    """ First assistant professor location """
    return person.first_asst_job_location

def get_current_location(person):
    """ Current location """ 
    return person.place

faculty = asst_faculty
location_func = get_asst_location  # Use their current or first assistant professorship location.

# Sanity check: does the number of topics line up with the number in the peoples' topic distributions?
found = False
for person in faculty:
    if 'topic_dist' in person:
        actual_num_topics = len(person['topic_dist'])
        found = True
        break    
if not found:
    raise ValueError("Couldn't determine the number of topics...")
else:
    if actual_num_topics != num_topics:
        raise ValueError("Number of topics does not match the number of topic descriptions!")

# Compute the "average" distribution
average_distribution = np.zeros(num_topics, dtype=float)
for person in faculty:
    if 'topic_dist' in person:
        average_distribution += person['topic_dist']
average_distribution /= average_distribution.sum()
```

```python
fig, ax = plt.subplots(figsize=(6,4))

selected = 'Average distribution'
ax.barh(xrange(num_topics), average_distribution, color=DARK_COLOR, edgecolor='None', align='center')
ax.set_yticks(xrange(num_topics))
ax.set_yticklabels(topic_descriptions, fontsize=LABEL_SIZE)
pad = 0.5
ax.set_ylim(-pad,num_topics-pad)
ax.tick_params(axis='both', which='major', labelsize=LABEL_SIZE, pad=7)
ax.set_title(selected, fontsize=LABEL_SIZE)

hide_right_top_axis(ax)

plt.show()
```

## Weighted Contributions

```python
lmbda = 10.452  # Average number of papers in all faculty
for f in faculty:
    if 'dblp_z' in f:
        f['dblp_w'] = f['dblp_z']*np.sqrt(lmbda) + lmbda
```

```python
# Get topic distribution at each institution
school_topics = {place:np.zeros(num_topics, dtype=float) for place in inst}
for place in inst:
    inst[place]['num_hires'] = 0
    inst[place]['avg_dblp_z'] = 0
    inst[place]['med_dblp_z'] = 0
    inst[place]['dblp_zs'] = []
    inst[place]['2011_counts'] = []

for person in faculty:
    place = location_func(person)
    if 'topic_dist' in person and place in school_topics:
        school_topics[place] += person['topic_dist']*(int(person['dblp_n_2011'])+1)
        inst[place]['num_hires'] += 1
        inst[place]['dblp_zs'].append(person['dblp_z'])
        inst[place]['2011_counts'].append(float(person['dblp_n_2011']))

for place in inst:
    if inst[place]['num_hires'] > 0:
        inst[place]['dblp_zs']
        inst[place]['avg_dblp_z'] = np.mean(inst[place]['dblp_zs'])
        inst[place]['med_dblp_z'] = np.median(inst[place]['dblp_zs'])
        inst[place]['2011_papers'] = np.mean(inst[place]['2011_counts']) #/ len(inst[place]['2011_counts'])
```
