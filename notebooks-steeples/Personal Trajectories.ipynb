```python
fig, ax = plt.subplots(figsize=(6,3.5))
years_after = 15
men = np.zeros(years_after, dtype=float)
women = np.zeros(years_after, dtype=float)
num_men = 0
num_women = 0
men_trajs = []
women_trajs = []

for person in asst_faculty:
    if 'dblp_pubs' not in person:
        continue 
        
        
    start = int(person.first_asst_job_year)
    if start >= 2002:
        continue
    
    year_diffs = []
    for pub in person['dblp_pubs']:
        if 'year' in pub and pub['year'] >= start:
            year_diffs.append(pub['year'] - start)

    num_pubs = float(len(year_diffs))
    pub_counts = Counter(year_diffs)
    last_pub = 2015 - start
    pub_traj = [pub_counts[y] for y in xrange(years_after)]
    
    if person.is_female:
        women += pub_traj
        women_trajs.append(pub_traj)
        num_women += 1 
        color = FEMALE_COLOR
    else:
        men += pub_traj
        men_trajs.append(pub_traj)
        num_men += 1
        color = ALMOST_BLACK
    
men_trajs = np.array(men_trajs)
women_trajs = np.array(women_trajs)

sig = []
for i in xrange(years_after):
    try:
        s, pval = mannwhitneyu(men_trajs[:,i], women_trajs[:,i])
        if pval < 0.05:
            sig.append(i)
    except:
        pass
    
ax.set_xticks(xrange(years_after))
men_avg = men / num_men
women_avg = women / num_women

for i in sig:
    ax.plot([i,i], [0, women_avg[i]], ':', color=ALMOST_BLACK)
   
ax.plot(men_avg, '-o', color=ALMOST_BLACK, markeredgecolor='w', linewidth=LINE_WIDTH, clip_on=False, label="Men")   
ax.plot(women_avg, '-o', color=FEMALE_COLOR, markeredgecolor='w',linewidth=LINE_WIDTH, clip_on=False, label="Women")   

ax.set_xlabel('Years after hire', fontsize=LABEL_SIZE)
ax.set_ylabel('Average publications per year', fontsize=LABEL_SIZE)
ax.tick_params(axis='both', which='major', labelsize=LABEL_SIZE, pad=7)

fake_line_all = Line2D(range(1), range(1), color=ALMOST_BLACK, marker='o',markersize=15, linestyle='None', markeredgecolor='w')
fake_line_f = Line2D(range(1), range(1), color=FEMALE_COLOR, marker='o',markersize=15, linestyle='None', markeredgecolor='w')
plt.legend((fake_line_all,fake_line_f),('Men','Women'), numpoints=1, loc='upper right', frameon=False, fontsize=LABEL_SIZE)

ax.set_xlim(-1, years_after)
hide_right_top_axis(ax)

plt.show()
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), sp.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return m, m-h, m+h

def median_confidence_interval(data):
    vals = np.sort(data)
    n = float(len(data))
    x = (1.96*np.sqrt(n))/2
    lower_ind = int(n-np.ceil((n/2)-x))
    upper_ind = int(n-np.floor(1+(n/2)+x))
    med = np.median(data)
    upper = vals[upper_ind]
    lower = vals[lower_ind]
    return med, lower, upper
```
