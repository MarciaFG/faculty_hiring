```python
def parse_doc_topics(doc_topics_file):
    
    # Get the last line, number of documents
    with open(doc_topics_file, 'rb') as fh:
        first = next(fh)
        offset = -100
        while True:
            fh.seek(offset, 2)
            lines = fh.readlines()
            if len(lines)>1:
                last = lines[-1]
                break
            offset *= 2
    
    pieces = last.split()
    num_docs = int(pieces[0]) + 1
    num_topics = (len(last.split()) - 2) / 2
    num_pieces = len(pieces)
    doc_topics = np.zeros((num_docs, num_topics), dtype=float)
    filenames = []

    for line in open(doc_topics_file, 'rU'):
        if line.startswith('#'):
            continue  # Skip comments
        pieces = line.strip().split()
        doc = int(pieces[0])
        filenames.append(pieces[1])
        
        for i in xrange(2, num_pieces, 2):
            doc_topics[doc][int(pieces[i])] = float(pieces[i+1])
            
    return filenames, doc_topics
```

## Average profile

```python
fig, ax = plt.subplots(figsize=(6, 4))

v_avg = np.zeros(num_topics, dtype=float)

for v in doc_topics:
    v_avg += np.sort(v)[::-1]
v_avg /= num_docs
    
ax.bar(xrange(num_topics), height=v_avg, align='center', color='0.8')
xlim_pad = 0.5
ax.set_xticks(xrange(num_topics))
ax.set_xlim((-xlim_pad, doc_topics.shape[1]-xlim_pad))
ax.set_xlabel('Sorted topic, 0 is most probable', fontsize=LABEL_SIZE)
ax.set_ylabel('Average probability', fontsize=LABEL_SIZE)
plt.show()  
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

v_avg = np.zeros(num_topics, dtype=float)

for v in doc_topics:
    v_avg += np.cumsum(np.sort(v)[::-1])
v_avg /= num_docs
    
ax.bar(xrange(num_topics), height=v_avg, align='center', color='0.8')
xlim_pad = 0.5
ax.set_xticks(xrange(num_topics))
ax.set_xlim((-xlim_pad, doc_topics.shape[1]-xlim_pad))
ax.set_xlabel('Sorted topic, 0 is most probable', fontsize=LABEL_SIZE)
ax.set_ylabel('Average cumulative probability', fontsize=LABEL_SIZE)
ax.set_ylim((0.5, 1.0))
plt.show()  
```
