```python
def make_trajectory_plot(ax1, school, rank=None, ax2size=1.2, decades=False, 
                         left_end=False, right_end=False, middle=False, ylim=None, show_dots=True,
                         show_right_ticks=True):
    color1 = ALMOST_BLACK
    color2 = FEMALE_COLOR

    divider = make_axes_locatable(ax1)
    ax2 = divider.append_axes("right", size=ax2size, pad=0, sharey=ax1)
    axs = [ax1, ax2]
    max_women = 0
    final_counts = []

    # Plot simulated trajectories
    for t in xrange(num_iterations):
        trajectory = np.cumsum(women_per_year[school][t])
        ax1.plot(year_range, trajectory, color=color2, linewidth=LINE_WIDTH, alpha=0.05)

        if trajectory[-1] > max_women:
            max_women = trajectory[-1]
        final_counts.append(trajectory[-1])

    # Plot the actual trajectory
    trajectory = np.cumsum(actual_women_per_year[school])
    ax1.plot(year_range, trajectory, color=color1, linewidth=LINE_WIDTH+1)
    
    # Add legend
    if show_dots:
        fake_line_1 = Line2D(range(1), range(1), color=color2, marker='o',markersize=15, linestyle='None', markeredgecolor='w')
        fake_line_2 = Line2D(range(1), range(1), color=color1, marker='o',markersize=15, linestyle='None', markeredgecolor='w')
        leg = ax1.legend((fake_line_1,fake_line_2),('Simulated','Actual'), numpoints=1, loc='upper left', 
                         frameon=False, fontsize=LABEL_SIZE+3, title=' ')
        leg_title = leg.get_title()
        #leg_title.set_position((15, 0))
        leg_title.set_ha("left")
        plt.setp(leg_title,fontsize=LABEL_SIZE+4)
    
    trans = ax1.get_xaxis_transform() # x in data untis, y in axes fraction
    if rank is not None:
        ann = ax1.annotate('%s, Rank #%d' % (school, rank), xy=(1972.75, 0.89), xycoords=trans, fontsize=LABEL_SIZE+4)
    else:
        ann = ax1.annotate(school, xy=(1972.75, 0.89), xycoords=trans, fontsize=LABEL_SIZE+4)

    ax1.spines["top"].set_visible(False)
    ax1.get_xaxis().tick_bottom()
    ax2.spines["top"].set_visible(False)
    ax2.get_xaxis().tick_bottom()
    
    y_pad = 0.5
    upper_y = np.max([max_women + 1, trajectory[-1] + 1])
    if ylim is not None:
        upper_y = ylim
        
    ax1.set_xlim(year_range[0], year_range[-1])
    ax1.set_ylim(-y_pad, upper_y-y_pad)
    ax2_x = arange(0, upper_y)
    double_digits = upper_y > 9
    if double_digits:
        ax2_x_ticks = arange(0, upper_y, 2)
    else:
        ax2_x_ticks = ax2_x
        
    ax2_y = np.array([sum(x == final_counts) for x in ax2_x], dtype=float)
    ax2_y /= np.sum(ax2_y)
    ax2.barh(ax2_x, ax2_y, color=color2, edgecolor='None', align='center', height=BAR_WIDTH)

    ind = int(where(ax2_x == trajectory[-1])[0])
    ax2.barh(ax2_x[ind], ax2_y[ind], edgecolor='None', align='center', color=color1, height=BAR_WIDTH)

    if decades:
        ax1.set_xticks([1970, 1980, 1990, 2000, 2010])
        ax1.set_xticklabels(["'70", "'80", "'90", "'00", "'10"])
    
    if left_end:
        ax1.set_ylabel('Cumulative female hires', fontsize=LABEL_SIZE+4)
    #else:
    #    ax1.set_yticklabels([])
        
    ax2.set_xticks([])
    ax2.set_yticks(ax2_x_ticks)
    ax1.set_yticks(ax2_x_ticks)
    ax2.set_xlim((0, np.max(ax2_y)+.05))
    ax2.set_ylim(-y_pad, upper_y-y_pad)
    if right_end or left_end:
        ax2.set_yticks(ax2_x_ticks)
        ax2.set_yticklabels(ax2_x_ticks)
        if right_end:
            ax2.set_ylabel('Distribution of final counts', fontsize=LABEL_SIZE+4, labelpad=10)

        if left_end:
            ax2.tick_params(axis='y', pad=20)
        else:
            ax1.tick_params(axis='y', pad=20)
    else:
        ax2.set_yticklabels([])
        
    if not double_digits:
        ax1.yaxis.labelpad = 12
        ax2.yaxis.labelpad = 12
    ax2.yaxis.tick_right()
    ax2.yaxis.set_label_position("right")
    
    
    for i in xrange(2):
        axs[i].tick_params(axis='both', which='major', labelsize=TICK_SIZE+2)
        axs[i].tick_params(axis='x', pad=XLABEL_PAD)
```

```python
def weighted_moving_average(x,y,step_size=0.01,width=1):
    # http://stackoverflow.com/questions/18517722/weighted-moving-average-in-python
    bin_centers  = np.arange(np.min(x),np.max(x)-0.5*step_size,step_size)+0.5*step_size
    bin_avg = np.zeros(len(bin_centers))

    #We're going to weight with a Gaussian function
    def gaussian(x,amp=1,mean=0,sigma=1):
        return amp*np.exp(-(x-mean)**2/(2*sigma**2))

    for index in range(0,len(bin_centers)):
        bin_center = bin_centers[index]
        weights = gaussian(x,mean=bin_center,sigma=width)
        bin_avg[index] = np.average(y,weights=weights)

    return (bin_centers,bin_avg)
```

```python
def weighted_moving_average(x,y,step_size=0.05,width=1):
    bin_centers  = np.arange(np.min(x),np.max(x)-0.5*step_size,step_size)+0.5*step_size
    bin_avg = np.zeros(len(bin_centers))

    #We're going to weight with a Gaussian function
    def gaussian(x,amp=1,mean=0,sigma=1):
        return amp*np.exp(-(x-mean)**2/(2*sigma**2))

    for index in range(0,len(bin_centers)):
        bin_center = bin_centers[index]
        weights = gaussian(x,mean=bin_center,sigma=width)
        bin_avg[index] = np.average(y,weights=weights)

    return (bin_centers,bin_avg)
```

```python
def get_dist(school):
    counts = Counter(np.sum(women_per_year[school], axis=1))
    p = np.array([counts[i] for i in xrange(int(max(counts))+1)], dtype=float)
    p /= p.sum()
    return p

def get_actual(school):
    return np.sum(actual_women_per_year[school])
```
