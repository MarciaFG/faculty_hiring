```python
def parse_stats_file(stats_file):
    stats = {}
    for line in open(stats_file, 'rU'):
        pieces = line.strip().split(':')
        if len(pieces) == 2:
            key = pieces[0]
            val = custom_cast(pieces[1])
            if key in stats:
                stats[key].append(val)
            else:
                stats[key] = [val]
    return stats

stats_dicts = {k:parse_stats_file(f) for k, f in zip(stats_desc, stats_files)}
compare_these = ['MEAN_GEODESIC', 'UW_RECIP', 'SELF_HIRES', 'SAME_REGION', 'AVG_CLUSTER', 'W_RECIP']
```

```python
def circle_legend(ax, colors, names, loc='upper left', markersize=15, ncol=1, frameon=False):
    circles = [Line2D(range(1), range(1), color=c, marker='o',markersize=markersize, linestyle='None', markeredgecolor='w')
               for c in colors]
    ax.legend(circles, names, numpoints=1, loc=loc, frameon=frameon, fontsize=LABEL_SIZE, ncol=ncol)
    if frameon:
        legend = ax.get_legend()
        frame = legend.get_frame()
        frame.set_facecolor('w')
        frame.set_edgecolor('w')
```

## Model Comparisons

```python
for key in compare_these:
    fig, ax = plt.subplots(figsize=(6,4))
    min_val = np.inf
    max_val = -np.inf

    for i, tag in enumerate(stats_desc):
        if tag != actual_tag:
            max_val = np.max([np.max(stats_dicts[tag][key]), max_val])
            min_val = np.min([np.min(stats_dicts[tag][key]), min_val])
        else:
            actual_val = stats_dicts[tag][key][0]

    if actual_val > max_val:
        max_val = actual_val + 0.25*(actual_val-max_val)
    if actual_val < min_val:
        min_val = actual_val - 0.25*(min_val-actual_val)

    x_plot = linspace(min_val, max_val, 1000)[:, newaxis]
    for i, tag in enumerate(stats_desc):
        if tag != actual_tag:
            vals = np.array(stats_dicts[tag][key])[:, newaxis]
            bw = 4*1.06*np.std(vals)*(len(vals)**(-1.0/5)) 
            mdl = KernelDensity(kernel='gaussian', bandwidth=bw)
            mdl.fit(vals)
            log_dens = mdl.score_samples(x_plot)
            ax.fill_between(x_plot[:, 0], exp(log_dens), 0, alpha=0.1, lw=1, color='0.4')
            ax.plot(x_plot[:, 0], exp(log_dens), lw=4, label=tag, color=colors[i])

    ylims = ax.get_ylim()
    ax.plot([actual_val, actual_val], ylims, '--', color=ALMOST_BLACK)
    ax.set_xlim(min_val, max_val)
    hide_right_top_axis(ax)
    circle_legend(ax, colors[:-1], stats_desc[:-1], loc='upper right', ncol=1, frameon=True)

    ax.set_yticks([])
    ax.set_xlabel(key_descriptions[key], fontsize=LABEL_SIZE)
    ax.set_ylabel(kde_ylabel, fontsize=LABEL_SIZE)

    plt.tight_layout()
    plt.savefig(os.path.join(figures_dir, 'STATS_%s.pdf' % key))
    fig.clf()
    plt.close()
```

```python
def format_number(number, precision):
    return "%.*f" % (precision, number)

print '''\\begin{table*}
\\centering
\\caption{Model checking (auto-generated.)}'''
display_tags = [actual_tag, 'Flat', 'Step', 'Sigmoid']
header_display_tags = [t.lower() for t in display_tags]
num_display_models = len(display_tags) - 1
line_end = '\\\\ \\hline'
display_keys = ['MEAN_GEODESIC', 'AVG_CLUSTER', 'W_RECIP', 'UW_RECIP', 'SELF_HIRES', 'SAME_REGION']
convert_perc = [False,          False,          True,      True,      True,          True]
num_digits =   [2,               2,             2,         2,         2,             2 ]

print '\\begin{tabular}{l|c|ccc}'
print '& & \\multicolumn{3}{c}{model $f$} \\\\' 
#print '\\begin{tabular}{r|c%s} \\hline' % ('|l'*(len(display_tags)-1))
#print '\\multicolumn{2}{c|}{} & \\multicolumn{%d}{c}{placement model}' % num_display_models, line_end
print ' & '.join([''] + header_display_tags), line_end

for k, key in enumerate(display_keys):
    values = []
    numbers = []
    for tag in display_tags:
        if tag == actual_tag:
            if convert_perc[k]:
                actual = stats_dicts[tag][key][0] * 100
            else:
                actual = stats_dicts[tag][key][0]
            values.append(format_number(actual, num_digits[k]))
            
        else:
            if convert_perc[k]:
                temp = 100*np.array(stats_dicts[tag][key])
                avg = np.mean(temp)
                std = np.std(temp)
            else:
                avg = np.mean(stats_dicts[tag][key])
                std = np.std(stats_dicts[tag][key])
            numbers.append(avg)
            values.append('%s $\\pm$ %s' % (format_number(avg, num_digits[k]), format_number(std, num_digits[k])))
            
    if convert_perc[k]:
        desc = key_descriptions[key].replace('Fraction', 'Percentage')
    else:
        desc = key_descriptions[key]
        
    diff = np.inf
    highlight = -1
    for i, val in enumerate(numbers):
        if np.abs(actual - val) < diff:
            highlight = i+1
            diff = np.abs(actual - val)
            
    value = values[highlight]
    #pieces = value.split()
    #pieces[0] = '\\textbf{%s}' % pieces[0]
    #values[highlight] = ' '.join(pieces)
    values[highlight] = '\\textbf{%s}' % value
    print ' & '.join([desc.lower()] + values), line_end

print '\\end{tabular}'
print '\\label{table:model:checking}'
print '\\end{table*}'
```
